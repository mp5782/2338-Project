```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./data/aggregate_mobility_05_04.csv")
data.head()
```

```python
def get_label(df,col):
    '''
    Create a label to determine the change in mobility:
    if mobility[t+1] > mobility[t]: label_mob = 1
    else 0
    '''
    data_label = list(df[col])
    temp = data_label.copy()
    for i in range(len(data_label)-1):
        temp[i] = data_label[i+1] - data_label[i]
    temp[-1] = temp[-2]
    df['label_'+col] = temp
    df['label_'+col] = np.where(df['label_'+col]<0,1,0)
    return df
```

```python
data_class_mobility = get_label(data,'Mobility')
data_class_mobility.head()
```

```python
data_class_case = get_label(data,'confirmed_infections')
data_class_case.head()
```
